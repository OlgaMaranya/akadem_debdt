```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Создание DataFrame для данных
data_normal = {
    'Эксперимент': ['Нормальные условия'] * 10,
    'Операция': range(1, 11),
    'Всего отправлений': [105, 102, 100, 98, 97, 101, 99, 103, 96, 104],
    'Средняя задержка транзакции (мс)': [8.9, 9.3, 9.1, 9.6, 9.4, 9.0, 9.8, 8.7, 9.2, 8.8],
    'Потенциальная загрузка каналов (mb)': [2.21, 2.15, 2.1, 2.06, 2.03, 2.12, 2.08, 2.17, 2.0, 2.19]
}

data_exp1 = {
    'Эксперимент': ['Эксперимент 1'] * 10,
    'Операция': range(1, 11),
    'Всего отправлений': [77, 78, 82, 76, 78, 73, 84, 75, 81, 87],
    'Средняя задержка транзакции (мс)': [12.49, 12.19, 12.4, 10.37, 11.77, 11.04, 8.76, 11.51, 10.84, 9.8],
    'Потенциальная загрузка каналов (mb)': [1.62, 1.64, 1.72, 1.6, 1.64, 1.53, 1.76, 1.57, 1.7, 1.83]
}

data_exp2 = {
    'Эксперимент': ['Эксперимент 2'] * 10,
    'Операция': range(1, 11),
    'Всего отправлений': [78, 83, 75, 80, 84, 79, 81, 81, 72, 78],
    'Средняя задержка транзакции (мс)': [9.01, 9.76, 8.4, 8.93, 10.21, 10.38, 10.64, 10.49, 9.1, 8.35],
    'Потенциальная загрузка каналов (mb)': [1.64, 1.74, 1.57, 1.68, 1.76, 1.66, 1.7, 1.7, 1.51, 1.64]
}

data_exp3 = {
    'Эксперимент': ['Эксперимент 3'] * 10,
    'Операция': range(1, 11),
    'Всего отправлений': [75, 77, 85, 78, 80, 81, 79, 75, 80, 82],
    'Средняя задержка транзакции (мс)': [10.73, 10.35, 9.67, 9.83, 9.21, 9.33, 9.51, 9.29, 11.85, 9.44],
    'Потенциальная загрузка каналов (mb)': [1.57, 1.62, 1.78, 1.62, 1.68, 1.7, 1.66, 1.57, 1.68, 1.7]
}

# Объединение всех данных
df_normal = pd.DataFrame(data_normal)
df_exp1 = pd.DataFrame(data_exp1)
df_exp2 = pd.DataFrame(data_exp2)
df_exp3 = pd.DataFrame(data_exp3)
df = pd.concat([df_normal, df_exp1, df_exp2, df_exp3], ignore_index=True)

# Настройка стиля
sns.set(style="whitegrid")

# 1. Линейные графики для сравнения динамики параметров
plt.figure(figsize=(14, 8))
plt.subplot(2, 1, 1)
sns.lineplot(data=df, x='Операция', y='Средняя задержка транзакции (мс)', hue='Эксперимент', marker='o')
plt.title('Динамика средней задержки транзакции')
plt.xlabel('Операция')
plt.ylabel('Задержка (мс)')

plt.subplot(2, 1, 2)
sns.lineplot(data=df, x='Операция', y='Потенциальная загрузка каналов (mb)', hue='Эксперимент', marker='o')
plt.title('Динамика потенциальной загрузки каналов')
plt.xlabel('Операция')
plt.ylabel('Загрузка (mb)')

plt.tight_layout()
plt.show()

# 2. Корреляционный график между задержкой и загрузкой каналов
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Средняя задержка транзакции (мс)', y='Потенциальная загрузка каналов (mb)', hue='Эксперимент')
plt.title('Корреляция между задержкой и загрузкой каналов')
plt.xlabel('Средняя задержка транзакции (мс)')
plt.ylabel('Потенциальная загрузка каналов (mb)')
plt.show()

# 3. Гистограмма для сравнения средних значений параметров
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Эксперимент', y='Средняя задержка транзакции (мс)', ci=None, palette='viridis')
plt.title('Сравнение средней задержки транзакции между экспериментами')
plt.xlabel('Эксперимент')
plt.ylabel('Средняя задержка (мс)')
plt.show()
```
