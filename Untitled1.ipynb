```python
import dash
from dash import dcc, html, Input, Output
import plotly.express as px
import pandas as pd
import numpy as np

# Генерируем данные
df = pd.DataFrame({
    'x': np.random.rand(50),
    'y': np.random.rand(50),
    'category': np.random.choice(['A', 'B', 'C'], 50)
})

# Создаем приложение
app = dash.Dash(__name__)

# Цветовая схема — единая для всех графиков
COLOR_MAP = {
    'A': '#1f77b4',  # Синий
    'B': '#ff7f0e',  # Оранжевый
    'C': '#2ca02c'   # Зелёный
}

# Порядок категорий
CATEGORY_ORDER = ['A', 'B', 'C']

# Основной график (маленький), передаём category как customdata
fig_scatter = px.scatter(
    df,
    x='x',
    y='y',
    color='category',
    custom_data=['category'],
    category_orders={'category': CATEGORY_ORDER},  # ← Устанавливаем порядок
    color_discrete_map=COLOR_MAP                   # ← Устанавливаем цвета
)

fig_scatter.update_layout(height=300, margin=dict(l=20, r=20, t=40, b=20))

# Макет
app.layout = html.Div([
    html.H2("Зависимые графики", style={'textAlign': 'center'}),
    
    # Основной график
    dcc.Graph(
        id='main-scatter',
        figure=fig_scatter,
        style={'width': '100%', 'height': '300px'}
    ),
    
    # Всплывающий график
    html.Div(id='hover-detail', 
             children="Наведите на точку",
             style={
                 'width': '100%',
                 'height': '250px',
                 'marginTop': '10px',
                 'backgroundColor': '#f9f9f9',
                 'border': '1px solid #ddd',
                 'padding': '10px'
             })
    
], style={
    'width': '100%',
    'padding': '10px',
    'box-sizing': 'border-box',
    'overflow': 'hidden'
})

# Callback для второго графика
@app.callback(
    Output('hover-detail', 'children'),
    Input('main-scatter', 'hoverData')
)
def display_hover_data(hoverData):
    if hoverData is None:
        return "Наведите на точку"

    try:
        category = hoverData['points'][0]['customdata'][0]
    except (KeyError, IndexError):
        return "Данные недоступны"

    # Генерация данных по категории
    if category == 'A':
        y_values = [1, 2, 3]
    elif category == 'B':
        y_values = [3, 2, 1]
    else:  # C
        y_values = [2, 3, 1]

    detail_df = pd.DataFrame({
        'x': [1, 2, 3],
        'y': y_values
    })

    fig_line = px.line(detail_df, x='x', y='y', title=f"Детали: {category}")
    fig_line.update_traces(line_color=COLOR_MAP[category])  # ← Цвет из COLOR_MAP
    fig_line.update_layout(height=230, margin=dict(l=20, r=20, t=30, b=20))

    return dcc.Graph(figure=fig_line)

# Запуск
if __name__ == '__main__':
    app.run(debug=True, mode='inline', height=650)
```
